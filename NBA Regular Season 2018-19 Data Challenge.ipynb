# What was the average age of player in the league?

```python
NBA_df2 = NBA_df['Age'].mean()
print(NBA_df2)
```

# What player scored the most points?

```python
NBA_df.sort_values(by = 'Points',ascending=False)
#James Harden
```

# What player had the most blocks during the season? Was it a post player (F/C)?

```python
NBA_df.describe(include = 'all')
NBA_df.sort_values(by = 'Blocks', ascending = False)
#Myles Turner has the most blocks  If a guard is considered a post player then yes if not then no.
```

# Based on the regular season, who had the best chance to win a title given their win percentage?

```python
ws_high = NBA_df.groupby('Team').agg({'WS': ['max',]})
print(ws_high.sort_values(by = ('WS', 'max'), ascending = False))
#Houston Rockets
```

# What player had the best 3-pt percentage? 

```python
# I'll assume that FG3%	is the same as the NBA's 3P%	3-Point Field Goal Percentage
NBA_df.describe(include = 'all')
NBA_df.head(10).sort_values(by = 'FG3%', ascending = False)
#Jaylen Adams
```

# Who played the most minutes during the season

```python
col = 'MP'
most_minutes =  NBA_df.loc[NBA_df[col].idxmax()]
print(most_minutes)
#Bradley Beal played the most minutes during the season
```

# What player given their player effiecency rating was the clutchest during the season?

```python
NBA_df.describe(include = 'all')
NBA_df.sort_values(by = 'PER', ascending = False)
#Zhou Qi has the highest PER
```

# What team had the youngest roster?

```python
grouped_single = NBA_df.reset_index()

grouped_single = NBA_df.groupby('Team').agg({'Age': ['mean',]})
print(grouped_single.sort_values(by = ('Age', 'mean'), ascending = True))
#The youngest team is the Chicago Bulls
```
